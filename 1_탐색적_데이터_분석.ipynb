* 제공 함수 생성
    * 변수 중요도를 시각화할 수 있는 함수를 제공합니다.
    * 입력 :
        * importance : 트리모델의 변수 중요도(예: model.feature_importances_)
        * names : 변수 이름 목록(예 : x_train.columns)
        * result_only  : 변수 중요도 순으로 데이터프레임만 return할지, 그래프도 포함할지 결정. False이면 결과 데이터프레임 + 그래프
        * topn : 중요도 상위 n개만 표시. all 이면 전체.
    * 출력 :
        * 중요도 그래프 : 중요도 내림차순으로 정렬
        * 중요도 데이터프레임 : 중요도 내림차순으로 정렬

```python
# 변수의 특성 중요도 계산하기
def plot_feature_importance(importance, names, result_only = False, topn = 'all'):
    feature_importance = np.array(importance)
    feature_name = np.array(names)

    data={'feature_name':feature_name,'feature_importance':feature_importance}
    fi_temp = pd.DataFrame(data)

    #변수의 특성 중요도 순으로 정렬하기
    fi_temp.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    fi_temp.reset_index(drop=True, inplace = True)

    if topn == 'all' :
        fi_df = fi_temp.copy()
    else :
        fi_df = fi_temp.iloc[:topn]

    #변수의 특성 중요도 그래프로 그리기
    if result_only == False :
        plt.figure(figsize=(10,20))
        sns.barplot(x='feature_importance', y='feature_name', data = fi_df)

        plt.xlabel('importance')
        plt.ylabel('feature name')
        plt.grid()

    return fi_df
```

### (1) 기본모델링 + 변수중요도 추출

```python
# 데이터 분할을 위한 전처리

target='Activity'
x = df_train.drop(target, axis=1)
y = df_train.loc[: , target]
```

#### 2) 모델링

#### 3) 변수 중요도 추출

```python
# 변수 중요도 추출(시각화로 나타내기)
full_importance_df = plot_feature_importance(
    importance=model.feature_importances_,
    names=x.columns,
    result_only=True,  # 시각화 안 하고 데이터만 가져오기
    topn='all'
)

max_importance = full_importance_df['feature_importance'].max()

top5_importance_df = plot_feature_importance(
    importance=model.feature_importances_,
    names=x.columns,
    result_only=True,  # 시각화도 함께 보고 싶다면 False
    topn=5               # 상위 5개만 보기
)

plt.figure(figsize=(8, 4))
plt.subplot(2,1,1)
sns.barplot(x='feature_importance', y='feature_name', data=top5_importance_df)
plt.title('Top 5 Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.xlim(0, max_importance)


bottom5_importance_df = full_importance_df.tail(5)

plt.subplot(2,1,2)
sns.barplot(x='feature_importance', y='feature_name', data=bottom5_importance_df)
plt.title('Bottom 5 Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.xlim(0, max_importance)

plt.grid()
plt.tight_layout()
plt.show()
```

* pd.merge를 이용해서 하나씩 합치는 작업을 반복 합니다.

#### 1) 이름변경

```python
display(full_importance_df.head()), display(full_importance_df2.head())
```

```python
# tip : rename을 써서 칼럼 이름 수정을 해야합니다!
full_importance_df = full_importance_df.rename(columns={'feature_importance': 'activity_importance'})
full_importance_df2 = full_importance_df2.rename(columns={'feature_importance': 'static_dynamic_importance'})

display(full_importance_df.head()), display(full_importance_df2.head())
```

#### 2) 합치기

```python
# merge 해서 2개를 합치는 작업하기
total_importance = pd.merge(full_importance_df, full_importance_df2, on='feature_name', how='outer')
```

```python
total_importance.sort_values(
    by=['activity_importance', 'static_dynamic_importance'],
    ascending=False
).to_csv(path + 'total_importance.csv')
```
